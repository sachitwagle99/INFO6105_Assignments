```python
from sklearn.feature_selection import chi2
cat=df.drop(['Call  Failure','Subscription  Length','Seconds of Use','Frequency of use','Frequency of SMS','Distinct Called Numbers','Age','Customer Value'],axis=1)
x=cat.drop("Churn",axis=1)
y=cat['Churn']
```

```python
chi2score=chi2(x,y)
sns.barplot(pd.Series(chi2score[1],index=x.columns).sort_values(ascending=False))
```

From above graphs, we can see that pretty much all numeric variables are linearly independent on the target variable (churn). Analysing the categorical data dependency on target variable, we can say that Age Group and Tariff Plan have no significant impact on target variable as their p values are > than 0.05(std p-value)

**12. Remove 1%, 5%, and 10% of your data randomly and impute the values back using at least 3 imputation methods. How well did the methods recover the missing values?  That is remove some data, check the % error on residuals for numeric data and check for bias and variance of the error.**


```python
print(f'1 % of data is {len(df)/100} rows')
print(f'5 % of data is {len(df)/20} rows')
print(f'10 % of data is {len(df)/10} rows')
```

```python
import random

oneperc = random.sample(range(len(df)), 32)
fiveperc = random.sample(range(len(df)), 158)
tenperc = random.sample(range(len(df)), 315)
```

```python
X1=X
X1.iloc[oneperc]=np.nan
X1.isna().sum()
X1cat=X1.drop(['Call  Failure','Subscription  Length','Seconds of Use','Frequency of use','Frequency of SMS','Distinct Called Numbers','Age','Customer Value'],axis=1)
X1num=X1.drop(X1cat.columns,axis=1)
```

```python

X1num.fillna(X1['Call  Failure'].mean(),inplace=True)
X1num.fillna(X1['Subscription  Length'].mean(),inplace=True)
X1num.fillna(X1['Seconds of Use'].mean(),inplace=True)
X1num.fillna(X1['Frequency of use'].mean(),inplace=True)
X1num.fillna(X1['Frequency of SMS'].mean(),inplace=True)
X1num.fillna(X1['Distinct Called Numbers'].mean(),inplace=True)
X1num.fillna(X1['Age'].mean(),inplace=True)
X1num.fillna(X1['Customer Value'].mean(),inplace=True)

X1cat.fillna(X1['Age Group'].mode()[0],inplace=True)
X1cat.fillna(X1['Tariff Plan'].mode()[0],inplace=True)
X1cat.fillna(X1['Status'].mode()[0],inplace=True)
X1cat.fillna(X1['Charge  Amount'].mode()[0],inplace=True)
X1cat.fillna(X1['Complains'].mode()[0],inplace=True)
```

```python
X1=pd.concat([X1num,X1cat],axis=1)
scaler=MinMaxScaler()
scaled=scaler.fit_transform(X1)
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size=0.25, random_state=101)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy : {accuracy}')
print(f'Confusion Matrix:\n {confusion_matrix(y_test,y_pred)}')
```

```python
X2=X
X2.iloc[fiveperc]=np.nan
X2.isna().sum()
X2cat=X2.drop(['Call  Failure','Subscription  Length','Seconds of Use','Frequency of use','Frequency of SMS','Distinct Called Numbers','Age','Customer Value'],axis=1)
X2num=X2.drop(X2cat.columns,axis=1)


X2num.fillna(X2['Call  Failure'].mean(),inplace=True)
X2num.fillna(X2['Subscription  Length'].mean(),inplace=True)
X2num.fillna(X2['Seconds of Use'].mean(),inplace=True)
X2num.fillna(X2['Frequency of use'].mean(),inplace=True)
X2num.fillna(X2['Frequency of SMS'].mean(),inplace=True)
X2num.fillna(X2['Distinct Called Numbers'].mean(),inplace=True)
X2num.fillna(X2['Age'].mean(),inplace=True)
X2num.fillna(X2['Customer Value'].mean(),inplace=True)

X2cat.fillna(X2['Age Group'].mode()[0],inplace=True)
X2cat.fillna(X2['Tariff Plan'].mode()[0],inplace=True)
X2cat.fillna(X2['Status'].mode()[0],inplace=True)
X2cat.fillna(X2['Charge  Amount'].mode()[0],inplace=True)
X2cat.fillna(X2['Complains'].mode()[0],inplace=True)

X2=pd.concat([X2num,X2cat],axis=1)
scaler=MinMaxScaler()
scaled=scaler.fit_transform(X2)
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size=0.25, random_state=101)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy : {accuracy}')
print(f'Confusion Matrix:\n {confusion_matrix(y_test,y_pred)}')
```

```python
X3=X
X3.iloc[tenperc]=np.nan
X3.isna().sum()
X3cat=X3.drop(['Call  Failure','Subscription  Length','Seconds of Use','Frequency of use','Frequency of SMS','Distinct Called Numbers','Age','Customer Value'],axis=1)
X3num=X3.drop(X2cat.columns,axis=1)


X3num.fillna(X3['Call  Failure'].mean(),inplace=True)
X3num.fillna(X3['Subscription  Length'].mean(),inplace=True)
X3num.fillna(X3['Seconds of Use'].mean(),inplace=True)
X3num.fillna(X3['Frequency of use'].mean(),inplace=True)
X3num.fillna(X3['Frequency of SMS'].mean(),inplace=True)
X3num.fillna(X3['Distinct Called Numbers'].mean(),inplace=True)
X3num.fillna(X3['Age'].mean(),inplace=True)
X3num.fillna(X3['Customer Value'].mean(),inplace=True)

X3cat.fillna(X3['Age Group'].mode()[0],inplace=True)
X3cat.fillna(X3['Tariff Plan'].mode()[0],inplace=True)
X3cat.fillna(X3['Status'].mode()[0],inplace=True)
X3cat.fillna(X3['Charge  Amount'].mode()[0],inplace=True)
X3cat.fillna(X3['Complains'].mode()[0],inplace=True)

X3=pd.concat([X3num,X3cat],axis=1)
scaler=MinMaxScaler()
scaled=scaler.fit_transform(X3)
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size=0.25, random_state=101)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy : {accuracy}')
print(f'Confusion Matrix:\n {confusion_matrix(y_test,y_pred)}')
```
